```python
train_df = pd.read_csv('../input/home-data-for-ml-course/train.csv')
test_df = pd.read_csv('../input/home-data-for-ml-course/test.csv')
```

# Model

```python
X_train = train_df[['OverallQual', 'OverallCond', 'LotArea']]
y_train = train_df['SalePrice']

model = linear_model.LinearRegression()
model.fit(X_train, y_train)

print('Coefficient:', model.coef_, '| Intercept:', model.intercept_)

# x = [[3300]]
# print(f"\ny = {model.coef_} * {x} + {model.intercept_} = {model.predict(x)}\n")

# areas = pd.DataFrame({'LotArea': [2700, 5000, 3900, 4300]})
# print(model.predict(areas))
```

```python
#### Predict the target variable using the fitted model
y_train_pred = model.predict(X_train)

# Calculate the R-squared score
r2 = r2_score(y_train, y_train_pred)

print(f"R-Squared: {r2:.4f}")
print(f"RMSE:      ${np.sqrt(mean_squared_error(y_train, y_train_pred)):,.2f}")
```

# Submission

```python
X_test = test_df[['OverallQual', 'OverallCond', 'LotArea']]

y_test_pred = model.predict(X_test)
```

```python
final_df = pd.DataFrame({
    'Id' : test_df['Id'], 
    'SalePrice' : y_test_pred
})

final_df.head()
```
